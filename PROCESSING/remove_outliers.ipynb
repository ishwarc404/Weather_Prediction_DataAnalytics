```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
df = pd.read_csv('IRENECSV.csv')
```

```python
sns.boxplot(x=df['pressure'])
```

```python
mean = df.loc[df['pressure']<2000,'pressure'].mean()
df["pressure"] = np.where(df["pressure"] >2000, mean,df['pressure'])
```

```python
sns.boxplot(x=df['pressure'])
```

```python
rslt_df = df.loc[df['month']==8]                         
rslt_df["pressure"] = np.where(rslt_df["pressure"] >2000, mean,rslt_df['pressure'])                                
fig = px.scatter(rslt_df, x='day', y='pressure')
fig.show()
```
